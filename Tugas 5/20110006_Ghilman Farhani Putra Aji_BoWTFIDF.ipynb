# Import Library

```python
import numpy as np
import pandas as pd
import os
from luwiji.text_proc import illustration
from nltk.tokenize import word_tokenize
from tqdm.auto import tqdm
from gensim.models import Word2Vec
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer, TfidfVectorizer
```

# Bag of Word

```python
text=pd.read_csv('data.csv', delimiter=',')
```

```python
text
```

```python
text=text["content"]
```

```python
bow = CountVectorizer()
bow_matrix = bow.fit_transform(text)
```

```python
bow_matrix
```

```python
pd.DataFrame(bow_matrix.toarray(), columns=bow.get_feature_names_out(), index=text)
```

# Inverse Document Frequency (IDF)

## Term Frequency - Inverse Document Frequency (TF-IDF)

```python
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(text)
```

```python
print(tfidf_matrix)
```

```python
tfidf.get_feature_names_out()
```

```python
pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf.get_feature_names_out(), index=text)
```
