Disini saya akan menggunakan library google play scraper untuk mendapatkan data aplikasi satu sehat

Langkah pertama adalah dengan mengimport semua library yang akan dibutuhkan

```python
from google_play_scraper import Sort, reviews
from google_play_scraper import app
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

kemudian, saya jalankan fungsi pada library google play scraper dengan memasukan variabel

kode nama aplikasi satu sehat

dengan bahasa indonesia

dan negara indonesia

kemudian saya sorting sesuai dengan yang terbaru

dengan jumlah data 1.000

kemudian untuk filter score yang diberikan adalah none(maksudnya adalah mengikutkan semua score yang ada didalam aplikasi tersebut)

```python
result, continuation_token = reviews(
    "com.telkom.tracencare",
    lang='id',
    country='id',
    sort=Sort.NEWEST,
    count=1000,
    filter_score_with=None
)
```

kemudian data yang diperoleh yang tadinya berbentuk array saya ubah menjadi data frame

```python
data=pd.DataFrame(np.array(result),columns=["review"])
data=data.join(pd.DataFrame(data.pop("review").tolist()))
```
