# Soal 2

<br>Lakukan pembersihan meliputi penghapusan punctuation, angka dan karakter yang tidak penting menggunakan menggunakan regex.
<br>Simpan hasilnya menjadi file csv/excel.


```python
#Jawaban
data=pd.DataFrame(np.array(result),columns=["review"])
data=data.join(pd.DataFrame(data.pop("review").tolist()))
data= data['content']
data
```

### menghapus typo

```python
import re
def replaceTypo(data):
    pola=re.compile(r'(.)\1{2,}',re.S)
    return pola.sub(r'\1',data)
data.apply(replaceTypo)
```

### menghapus angka

```python
number='01234567890'
def remove_number(text):
    remove=number
    return text.translate(str.maketrans('','',remove))
data.apply(lambda x: remove_number(x))
data
```

### lowercasing

```python
data=data.str.lower()
data
```

### tokenizing

```python
import nltk
from nltk import word_tokenize, sent_tokenize
word_token=[]
for i in data:
    token = word_tokenize(i)
    word_token.append(token)
```

```python
word_token
```

### menghapus tanda

```python
from string import punctuation
```

```python
token_hapus_tanda=[]
for i in word_token:
    loop = [token for token in i if token not in punctuation]
    token_hapus_tanda.append(loop)
```

```python
token_hapus_tanda
```

```python
token_hapus_tanda_csv=pd.DataFrame(token_hapus_tanda)
token_hapus_tanda_csv.to_csv("token_hapus_tanda_csv.csv")
```

b.Carilah kata-kata slangword yang ada dalam dataset Anda, dengan cara mencocokan dengan kamus KBBI (terlampir). 
<br>Simpan hasilnya dalam bentuk csv/excel.
<br> Tampilkan 100 kata slang yang Anda dapatkan dan tampilkan dalam bentuk Gunakan (bobot:25)

```python
#Jawaban
slang_words=pd.read_csv("slangword.csv")

words=''.join(data)
words=words.split()
output_words=[]
output_text=[]
for word in words:
    if word in slang_words:
        output_words.append(slang_words[word])
    else:
        output_words.append(word)
        output_text = " ".join(output_words)
print("Hasil: ", output_text)
```

```python
kata_slang=pd.DataFrame(output_words)
kata_slang.to_csv("kata_slang.csv")
output_words[0:99]
```
